##List Comprehension

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = []
for x in fruits:
  if 'a' in x:
    newlist.append(x)
print(newlist)

#syntax :newlist = [expression for item in iterable if condition == True]

newllist=[x for x in fruits if 'b' in x]
print(newllist)
```

###Iterable
The iterable can be any iterable object, like a list, tuple, set etc.


```python
newlist =[x for x in range(10)]
print(newlist)
newlist =[x for x in range(10) if x<5]
print(newlist)
```

###Expression: The expression is the current item in the iteration, but it is also the outcome, which you can manipulate before it ends up like a list item in the new list:



```python
newlist = [x.upper() for x in fruits]
print(newlist)
newlist=[ 'hello' for x in fruits]
print(newlist)
```

```python
newlist =[x if x != 'banana' else 'orange' for x in fruits]
print(newlist)
```

###Customize Sort Function


```python
#key=function
def myfunc(n):
  return abs(n-50)
thislist =[100,50,65,82,23]
thislist.sort(key=myfunc)
print(thislist)
```
